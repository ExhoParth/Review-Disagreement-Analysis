**Let's see the features included in our dataset**

```python
# Load JSON data
with open("/Users/viru/Documents/GitHub/Review-Disagreement-Analysis/dataset/train/B1euHOqi37.json", 'r') as file:
    data = json.load(file)

# Display the JSON data
print(data)
```

```python
df = json_normalize(data)
df.head()
```

```python
df.describe(include='O').T
```

```python
df.shape
```

```python
print("Features: ", df.columns)
```

```python
# Use json_normalize to flatten the dictionary column
flattened_df = pd.json_normalize(df['review_sentences'])

# Merge flattened columns back with the original DataFrame (if needed)
review_sent_df = df.drop(columns=['review_sentences']).join(flattened_df)

# Display the final DataFrame
review_sent_df.head()
```

```python
review_sent_df = review_sent_df[['review_id', 'sentence_index', 'text', 'suffix',
       'review_action', 'fine_review_action', 'aspect', 'polarity']]
review_sent_df.head()
```

**Feature in each review sentences above are:**
1. review_id
2. sentence_index
3. text --> given by the reviewer
4. suffix
5. review_action --> action type
6. fine_review_action
7. aspect --> what are the topic of review
8. polarity
